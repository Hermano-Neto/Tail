```python
# Uniformizando cabeçalho, deixe-o em caixa baixa (minúsculo)
df.columns = df.columns.str.lower()
df_copia = df.copy() 
df_copia.columns = df_copia.columns.str.lower()
df_copia.head()
df
```

```python
# Verifique valores únicos na coluna Publisher

print('Valores únicos na coluna Publisher:\n\n', df['publisher'].unique())
```

```python
# Deixe em minúsculo os nomes dos publishers, conte valores únicos e veja se houve alteração

df['publisher'] = df['publisher'].str.lower()
print('Quantidade de valores únicos na coluna Publisher:', len(df['publisher'].unique()),'\n')
print('Quantidade de valores por valor único:\n', df['publisher'].value_counts())
df
```

```python
# Verificando existência de valores duplicados

print('Quantidade de valores duplicados:', df.duplicated().sum())
# >>> Como não há valores/linhas duplicados não precisamos fazer nenhuma modificação aqui
```

```python
# Verificando existência de valores NAN

print("Quantidade de valores NAN por coluna:\n", df.isna().sum(), '\n')
print("Porcentagem de valores NAN por coluna:\n", (df.isna().sum()/df.shape[0]) * 100)
```

```python
# Verifique distribuições .hist() antes de tratar NaN

df.hist()
```

```python
# Utilizando média para preencher NaN na coluna critic count

df['critic_count'] = df['critic_count'].fillna(df['critic_count'].mean())
print("Quantidade de valores NAN na coluna critic_count:", df['critic_count'].isna().sum())
df
```

```python
# Verificando se houve mudança significativa na distribuição da variável com .hist()

#df.hist()
df['critic_count'].hist()
# >>> Se nota que houve mudança significativa!
```

OPS! Observe que quando a há muitos valores NaN (nesse caso, aproximadamente 50%), é preferível utilizar outras formas para preencher valores NaN, tais como substituindo por valores interpolados ou procurando valores já existentes na mesma coluna. Importante frisar que depende do tipo de dado a ser preenchido. Caso o dado fosse a temperatura da cidade de João Pessoa durante as manhãs do mês de março, que devem variar entre 28°C e 31°C, utilizar a média não causaria uma distorção.

```python
# Preenchendo NaN da coluna Critic Score com valores interpolados pelo método linear
df_copia['critic_score'] = df_copia['critic_score'].fillna(df_copia['critic_score'].interpolate(method = 'linear'))
df_copia['critic_score']
```

```python
# Verificando distribuição (.hist()) do Critic Score após interpolação
df_copia['critic_score'].hist()

```

```python
# Preenchendo NaN da coluna Critic Count com valores interpolados k-nn
df_copia['critic_count'] = df_copia['critic_count'].fillna(df_copia['critic_count'].interpolate(method = 'nearest'))
df_copia['critic_count']
```

```python
# Agrupando por nome e ano de lançamento
name_rearRelease = df_copia.groupby('name')['year_of_release'].unique()
name_rearRelease
```

```python
# Preenchendo NaN no ano de lançamento com .map() no groupby
plat_yearOfRele = df_copia.groupby('platform')['year_of_release'].unique()

df_copia['year_of_release'].fillna(df_copia['platform'].map(plat_yearOfRele), inplace = True)
df_copia.info()
```

```python
# Agrupando por nome e publisher
name_publisher = df_copia.groupby('name')['publisher'].unique()
name_publisher
```

```python
# Preenchendo NaN no publisher com .map()
plat_publisher = df_copia.groupby('platform')['publisher'].unique()

df_copia['publisher'].fillna(df_copia['platform'].map(plat_yearOfRele), inplace = True)
df_copia.info()
```
